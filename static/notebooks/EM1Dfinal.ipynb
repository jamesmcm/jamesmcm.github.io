Expectation Maximisation Algorithm: Mixture of Gaussians
===================================

Note that in this work $\boldsymbol \theta^t$ means the value of $\boldsymbol \theta$ at time $t$. $\ \mathcal{N} (x;\mu, \sigma)$ is the Probability Density Function of a Normal distribution with mean $\mu$ and variance $\sigma^2$.

In the code, sigma1 and sigma2 are actually $\sigma_j^2$ values

General idea of EM:
--------------------

The general idea is to maximise the log-likelihood by iteratively updating our cluster assignments (i.e. our best estimate of which Gaussian the data point came from, given the current estimate of their parameters), and then the parameters of these Gaussians given the assignments.

Definition of the likelihood (i.e. probability of the data given the model), but note we usually cannot marginalise over $\mathbf{Z}$ directly: 

$L(\boldsymbol \theta; \mathbf{X} ) = P(\mathbf{X} | \boldsymbol \theta ) = \sum_z P(\mathbf{X}, \mathbf{Z} | \boldsymbol \theta ) $

So instead we take this iterative updating approach.

Assigning Z (i.e. the cluster assignments in this example): 

$Q(\boldsymbol \theta | \boldsymbol \theta^t ) = \mathbb{E}_{Z|X,\theta^t}$ \left [ \log ( L(\boldsymbol \theta; \mathbf{X}, \mathbf{Z} ) \right ] $

Updating the parameters:

$\boldsymbol \theta ^{t+1} = \arg\max_\theta Q(\boldsymbol \theta | \boldsymbol \theta^t ) $


 
Background for this example (1D GMM):
---------------------------------------

Example taken from: http://en.wikipedia.org/wiki/Expectation%E2%80%93maximization_algorithm

Data points: $\mathbf{X} = (X_1 \dots X_n )$

Gaussian assignments: $\mathbf{Z} = (Z_1 \dots Z_n )$ note assignment probabilities are in $\mathbf{T}$

Prior on cluster assignment :
$P(Z_i=1)=\tau_1 , P(Z_i=2)=\tau_2 = 1 - \tau_1$   

All parameters: 
$\mathbf{\boldsymbol \theta} = ( \mathbf{ \boldsymbol \tau}, \mu_1, \mu_2, \sigma^2_1, \sigma^2_2 )$

Likelihood:
$L(\boldsymbol \theta; \mathbf{X}, \mathbf{Z}) = P(\mathbf{X}, \mathbf{Z} | \boldsymbol \theta) = \prod_{i=1}^n P(Z_i = z_i ) \  \mathcal{N} (x_i | (\mu_j, \sigma^2_j)) = \prod_{i=1}^n \sum_{j=1}^2 \mathbb{I} (z_i = j) \tau_j \   \mathcal{N} (x_i, \mu_j, \sigma^2_j)$

In exponential form (so we can take the log):

$L(\boldsymbol \theta ; \mathbf{X}, \mathbf{Z}) = \exp \left ( \sum_{i=1}^n  \sum_{j=1}^2 \mathbb{I}(Z_i=j) \left [ \log (\tau_j) - \log(\sigma_j )- \frac{1}{2\sigma^2_j} (x_i - \mu_j)^2 - \frac{d}{2} \log(2\pi) \right ] \right )$

Where $ \mathbb{I}(Z_i=j)$ is the indicator function (i.e. returns 1 iff. $Z_i = j$, else returns 0)

So log-likelihood, which we want to maximise: 

$\log ( L(\boldsymbol \theta ; \mathbf{X}, \mathbf{Z}) ) = \sum_{i=1}^n  \sum_{j=1}^2 \mathbb{I}(Z_i=j) \left [ \log (\tau_j) - \log(\sigma_j) - \frac{1}{2\sigma^2_j} (x_i - \mu_j)^2 - \frac{d}{2} \log(2\pi) \right ]$


Expectation step:
------------------
Gaussian assigment probabilities are in $\mathbf{T}$, this is what we wish to return in the E-step. 

$\mathbf{T}_{j,i}^t  = P(Z_i = j | X_i=x_i, \boldsymbol \theta^t) = \frac{\tau_j^t \ \mathcal{N}(x_i; \mu_j^t, \sigma_j^t)}{\tau_1^t \ \mathcal{N}(x_i; \mu_1^t, \sigma_1^t) + \tau_2^t \ \mathcal{N}(x_i; \mu_2^t, \sigma_2^t) }$

This can also be expressed in terms of $Q$:

$Q(\mathbf \theta | \theta^t ) = \mathbb{E} \left [ \log L(\boldsymbol \theta; \mathbf{X}, \mathbf{Z} ) \right ] = \mathbb{E} \left [ \log \prod_{i=1}^n L(\boldsymbol \theta; \mathbf{X}_i, \mathbf{Z}_i ) \right ] = \mathbb{E} \left [ \sum_{i=1}^n \log  L(\boldsymbol \theta; \mathbf{X}_i, \mathbf{Z}_i ) \right ] = \sum_{i=1}^n \mathbb{E} \left [  \log  L(\boldsymbol \theta; \mathbf{X}_i, \mathbf{Z}_i ) \right ]$

$Q(\boldsymbol \theta | \theta^t )= \sum_{i=1}^n \sum_{j=1}^2 \mathbf{T}_{j,i}^t \left [ \log \tau_j - \log ( \sigma_j ) - \frac{1}{2\sigma_j^2} (X_i - \mu_j) - \frac{d}{2} \log(2\pi ) \right ] $


Maximisation step:
------------------

In this step we update our estimates for all of the parameters:

First let's update our estimates for $\mathbf{\tau}$, the prior probabilities for assignment to the Gaussians. Recall that $\tau_1 + \tau_2 = 1$. The update rule is as follows:

$\tau^{t+1} = \arg\max_{\boldsymbol \tau} Q(\boldsymbol \theta | \boldsymbol \theta^t ) = \arg\max_\tau \left ( \left [ \sum_{i=1}^n \mathbf{T}_{i,1}^t \right ] \log \tau_1 + \left [ \sum_{i=1}^n T_{2,i}^t \right ] \log \tau_2 \right ) $

This has the same for as the Maximum Likelihood Estimate for the binomial distribution, so the update rule is:

$\tau_j^{t+1} = \frac{\sum_{i=1}^n \mathbf{T}_{j,i}^t }{\sum_{i=1}^n \left ( \mathbf{T}_{1,i}^t + \mathbf{T}_{2,i}^t \right )} = \frac{1}{n} \sum_{i=1}^n  \mathbf{T}_{j,i}^t$

We now wish to estimate $\mu_1$ and $\sigma_1$:

$(\mu_1^{t+1} , \sigma_1^{t+1} ) = \arg\max_{\mu_1, \sigma_1} Q ( \boldsymbol \theta | \boldsymbol \theta^t) = \arg\max_{\mu_1 , \sigma_1} \sum_{i=1}^n \mathbf{T}_{1,i}^t \left ( - \log(\sigma_1 ) - \frac{1}{2 \sigma_1 ^2 } (x_i-\mu_1)^2 \right )$

This has the same form as a weighted Maximum Likelihood Estimator for a normal distribution, so the update rules are as follows:

$\mu_1^{t+1} = \frac{\sum_{i=1}^n \mathbf{T}_{1,i}^t X_i}{\sum_{i=1}^n T_{1,i}^t }$

$\sigma_1^{t+1} = \frac{\sum_{i=1}^n \mathbf{T}_{1,i}^t \left (X_i - \mu_1^{t+1} \right ) ^2}{\sum_{i=1}^n \mathbf{T}_{1,i}^t}$

And equally the the second Gaussian (note the updates are independent):

$\mu_2^{t+1} = \frac{\sum_{i=1}^n \mathbf{T}_{2,i}^t X_i}{\sum_{i=1}^n T_{2,i}^t }$

$\sigma_2^{t+1} = \frac{\sum_{i=1}^n \mathbf{T}_{2,i}^t \left (X_i - \mu_2^{t+1} \right ) ^2}{\sum_{i=1}^n \mathbf{T}_{2,i}^t}$


We carry out these updates until the log-likelihood converges.

```python
import numpy as np
import pylab as pb
%pylab inline



def generate_data(m1,m2,s1,s2):
    global mu1, mu2, sigma1, sigma2
    mu1=m1
    mu2=m2
    sigma1=s1
    sigma2=s2
    #generate a mixture of a 1D gaussian - plot histogram?
    X=np.hstack([mu1 + np.sqrt(sigma1)*pb.randn(1000),mu2 + np.sqrt(sigma2)*pb.randn(1000)])
    return X


def plot_data(X):
    global mu1, mu2, sigma1, sigma2
    # the histogram of the data with histtype='step'
    n, bins, patches = pb.hist(X, 50, normed=1, histtype='stepfilled')
    pb.setp(patches, 'facecolor', 'g', 'alpha', 0.75)
    # add a line showing the expected distribution
    y = (pb.normpdf( bins, mu1, np.sqrt(sigma1)) + pb.normpdf( bins, mu2, np.sqrt(sigma2)))/2.0
    l = pb.plot(bins, y, 'k--', linewidth=1.5)

def loglikelihood(X,parameters,Z):
    ll=0
    for i in range(X.size):
        lt=0
        if Z[i]==0:
            lt+=np.log(parameters["tau1"])- 0.5*np.log(parameters["sigma1"]) - (1.0/(2*parameters["sigma1"]))*(X[i]-parameters["mu1"])**2 - np.log(2*np.pi)
        else:
            lt+=np.log(parameters["tau2"])- 0.5*np.log(parameters["sigma2"])-(1.0/(2*parameters["sigma2"]))*(X[i]-parameters["mu2"])**2 - np.log(2*np.pi)
            
        ll+=lt
    return ll


def estep(X,parameters):
    #return assignment probabilities T
    ts=[]
    for i in range(X.size):
        t1=(parameters["tau1"]*pb.normpdf( X[i], parameters["mu1"], np.sqrt(parameters["sigma1"])))/(parameters["tau1"]*pb.normpdf( X[i], parameters["mu1"], np.sqrt(parameters["sigma1"])) + parameters["tau2"]*pb.normpdf( X[i], parameters["mu2"], np.sqrt(parameters["sigma2"])))
        t2=(parameters["tau2"]*pb.normpdf( X[i], parameters["mu2"], np.sqrt(parameters["sigma2"])))/(parameters["tau1"]*pb.normpdf( X[i], parameters["mu1"], np.sqrt(parameters["sigma1"])) + parameters["tau2"]*pb.normpdf( X[i], parameters["mu2"], np.sqrt(parameters["sigma2"])))
        ts.append(np.array([t1,t2]))
        
    return np.vstack(ts)

def mstep(X,parameters, T):
    #return new parameters theta_t+1
    newparams={}
    newparams["tau1"]=1/(float(X.size)) * np.sum(T[:,0])
    newparams["tau2"]=1/(float(X.size)) * np.sum(T[:,1])
    
    newparams["mu1"]=np.dot(T[:,0],X)/float(np.sum(T[:,0]))
    newparams["mu2"]=np.dot(T[:,1],X)/float(np.sum(T[:,1]))

    newparams["sigma1"]=np.dot(T[:,0],np.square(X-newparams["mu1"]))/float(np.sum(T[:,0]))
    newparams["sigma2"]=np.dot(T[:,1],np.square(X-newparams["mu2"]))/float(np.sum(T[:,1]))
    
    return newparams

def plotpred(X, parameters):
    global mu1, mu2, sigma1, sigma2
    # the histogram of the data with histtype='step'
    n, bins, patches = pb.hist(X, 50, normed=1, histtype='stepfilled')
    pb.setp(patches, 'facecolor', 'g', 'alpha', 0.5)
    # add a line showing the expected distribution

    
    y = pb.normpdf( bins, parameters["mu1"], np.sqrt(parameters["sigma1"]))/2.0
    l = pb.plot(bins, y, 'k-', linewidth=1.5)
    y2=pb.normpdf( bins, parameters["mu2"], np.sqrt(parameters["sigma2"]))/2.0
    l2 = pb.plot(bins, y2, 'k-', linewidth=1.5)
    d = np.zeros(len(y))
    pb.fill_between(bins,y,where=y>=d, color='blue')
    pb.fill_between(bins,y2,where=y2>=d, color='red')
    
    y4=pb.normpdf( bins, mu1, np.sqrt(sigma1))/2.0
    l = pb.plot(bins, y4, 'k--', linewidth=1, color="magenta")
    y5=pb.normpdf( bins, mu2, np.sqrt(sigma2))/2.0
    l = pb.plot(bins, y5, 'k--', linewidth=1, color="magenta")    
    #y3 = (pb.normpdf( bins, mu1, np.sqrt(sigma1)) + pb.normpdf( bins, mu2, np.sqrt(sigma2)))/2.0
    #l = pb.plot(bins, y3, '-', linewidth=1, color="magenta")
    
    

def step(x):
    if x>=0.5:
        return 0
    else:
        return 1
    
vstep=np.vectorize(step)
```

```python
%pylab inline
#print generate_data()
X=generate_data(-2,3,5,2)
plot_data(X)
```

```python
parameters={"tau1":0.5, "tau2":0.5, "mu1":-5,"mu2":6,"sigma1":3, "sigma2":0.99}

pparams=[]
T=estep(X,parameters)
Z=vstep(T[:,0])

pparams.append((parameters,loglikelihood(X,parameters,Z)))
for i in range(50):
    T=estep(X,parameters)
    #print T
    newparams=mstep(X,parameters,T)
    #print newparams
    Z=vstep(T[:,0])
    print loglikelihood(X,parameters,Z)
    parameters=newparams
    pparams.append((parameters,loglikelihood(X,parameters,Z)))

```
